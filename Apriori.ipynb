```python
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd  
```

```python
#Importing the dataset  
# Data Preprocessing
dataset = pd.read_csv("C:/Users/Nehal/OneDrive/Desktop/ML R Models/Apriori/Market_Basket_Optimisation.csv",header = None)  
transactions=[]  
for i in range(0, 7501):  
    transactions.append([str(dataset.values[i,j])  for j in range(0,20)])
```

```python
from apyori import apriori  
rules= apriori(transactions= transactions, min_support=0.003, min_confidence = 0.2, min_lift=3, min_length=2, max_length=2)
```

```python
results= list(rules)  
results 
```

```python
for item in results:  
    pair = item[0]   
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])  
    print("Support: " + str(item[1]))  
    print("Confidence: " + str(item[2][0][2]))  
    print("Lift: " + str(item[2][0][3]))  
    print("=====================================")  
```

Putting the result well organised into a pandas Dataframe

```python
def inspect(results):
    lhs =[tuple(result[2][0][0])[0]for result in results]
    rhs =[tuple(result[2][0][1])[0]for result in results]
    supports = [result[1]for result in results]
    confidence = [result [2][0][2] for result in results]
    lifts = [result [2][0][3] for result in results]
    return list(zip(lhs, rhs, supports,confidence, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results),columns = ['Left Hand Side','Right Hand Side','Support','Confidence','lift'])
```
